# Data clean up and validation


```python
cars_file.duplicated().sum()
# I want to verify if there are duplicate values in the dataset. Too many duplicate values
```

```python
cars_file.drop_duplicates()
# Removing duplicates 
```

```python
# I want to check if there are missing values in the dataset
cars_file.isnull()
```

```python
cars_file.isnull().sum()
# confirmed that there is no missing missing value
```

```python
# cleaning the Price column
cars_file = cars_file[cars_file['Price'] != 'Not Priced']
cars_file['Price'] = cars_file['Price'].str.replace('$','').str.replace(',','')
cars_file['Price'] = cars_file['Price'].astype(int)
cars_file.head()

# Removing the $ sign and making the value as integer 
```

```python
# cleaning the Mileage columns
cars_file = cars_file[cars_file['Mileages'] != 'Mileage']
cars_file['Mileages'] = cars_file['Mileages'].str.replace('mi.','').str.replace(',','').astype(int)
cars_file.head()

# Cleaning the Mileages column and removing 'mi' in Mileages column and changing the value as Integer
```

```python
# cleaning Reviews column

cars_file['Reviews'] = cars_file['Reviews'].astype(str).str.replace(' reviews','').str.replace(',','').str.findall(r'[0-9]').apply(lambda x: ''.join(x)).astype(int)
cars_file.head()

# Cleaned the Reviews column and made it as integer for better analysis
```

```python
# Cleaning Car Names columns

cars_file['Cars'] = cars_file['Car Names'].str.split(' ').apply(lambda x: x[1])
cars_file['ManufactureYear'] = cars_file['Car Names'].str.split(' ').apply(lambda x: x[0])
cars_file.head()

# added Cars column to identify the model of the car
# added Manufacture Year column to identify the date it was made
```

```python
# Cleaning Used/Certified column

def used_cer(x):
    if 'Used' in x:
        return 'Used'
    elif 'Certified' in x:
        return 'Certified'
cars_file['used/certified'] = cars_file['used/certified'].apply(used_cer)

cars_file.head()
```

```python
# Cleaning price drop column

cars_file['price drop'] = cars_file['price drop'].str.split(' ').apply(lambda x: x[0]).str.replace('$','').str.replace(',','').astype(int)

# removing the word "price drop" in the column retaining the value as integer
```

```python
# Removing the Car Names column 
cars_file.drop('Car Names',axis = 1, inplace = True)

# renaming Cars column to Car Names
cars_file.rename(columns = {'Cars':'Car Names'}, inplace = True)

cars_file.head()
```

# Exploration of Data and Analysis

```python
# Identifying the cars with the most mileage and better pricing before 2018
cars2018below = cars_file[cars_file['ManufactureYear'] < '2019']
```

```python
cars2018below.head()
```

```python
# Changing the style for all graphs
plt.style.use('bmh')


# I want to show the best car with better mileage and prices of cars before 2018
sns.relplot(x='Mileages',y='Price', hue='Car Names', data=cars2018below, height = 8, aspect = 2).set(title='Cars 2018 and below')

# Saving the figure
plt.savefig("Cars2018below.jpeg")

plt.show()
```

```python
# I want to sort the data by its Manufacture year, price and Mileages

car2018sort = cars2018below.sort_values(["ManufactureYear","Price","Mileages"])
car2018sort.head()
```

```python
#changing the size of the graph to become readable
plt.figure(figsize = (30,10))

# I want to see the price of the car over its mileages
ax = plt.subplot()
ax.barh(car2018sort["Car Names"], car2018sort['Mileages'])
ax.barh(car2018sort["Car Names"], car2018sort["Price"], color="maroon")
ax.set_title("Miles and Price")

# Saving the figure
plt.savefig('2018MilesandPrice.jpg',quality = 50)
plt.show()
```

```python
# I want to show the changes in prices and mileages for cars per year
fig, ax = plt.subplots(figsize = (20,10))
ax.plot(car2018sort['ManufactureYear'], car2018sort['Price'], color='r')
ax.set_xlabel('Year of Manufacture')
ax.set_ylabel('Price of Car', color = 'red')
ax.tick_params('y', colors = 'red')
ax2=ax.twinx()
ax2.plot(car2018sort['ManufactureYear'],car2018sort['Mileages'], color = 'blue')
ax2.set_ylabel('Mileages per gallon', color = 'blue')
ax2.tick_params('y', colors = 'blue')

# Saving the figure
fig.savefig('ChangesinPriceandMileages.jpg', quality = 50)
plt.show()
```

```python
# Grouping the cars per year and plotting the chart against its price

car2018sort.groupby(['ManufactureYear', 'Car Names']).sum().unstack().plot(figsize=(10,10), kind='bar', y='Price', stacked = True, title = "Total Prices per Brand")

# Saving the figure
plt.savefig('PricesofcarsperBrand.jpg', quality = 50)
plt.show()
```

```python
# Identifying the cars after 2018
newercars = cars_file[cars_file['ManufactureYear'] > '2018']
newercars.head()
```

```python
# Verifying the best cars after 2018

sns.relplot(x='Mileages',y='Price', hue='Car Names', data=newercars,height = 8, aspect = 2).set(title="Newer Cars") 

# saving the graph
plt.savefig("Allnewcars.jpg", quality = 50)
plt.show()
```

```python
# Sorting the data for the newer cars

newercarssort = newercars.sort_values(["ManufactureYear", "Price"])
newercars.head()
```

```python
# Plotting the same analysis for newer cars
plt.figure(figsize = (30,10))
ax = plt.subplot()
ax.barh(newercarssort["Car Names"], newercarssort['Mileages'])
ax.barh(newercarssort["Car Names"], newercarssort["Price"], color="green")
ax.set_title("New Cars Miles and Price")

# Saving the graph
plt.savefig("NewCarsMandP.jpg", quality = 50)
plt.show()
```

```python
# Graphing the change of miles and prices per year for newer cars

fig, ax = plt.subplots(figsize = (20,10))
ax.plot(newercarssort['ManufactureYear'], newercarssort['Price'], color='maroon')
ax.set_xlabel('Year of Manufacture')
ax.set_ylabel('Price of Car', color = 'maroon')
ax.tick_params('y', colors = 'red')
ax2=ax.twinx()
ax2.plot(newercarssort['ManufactureYear'],newercarssort['Mileages'], color = 'orange')
ax2.set_ylabel('Miles per gallon', color = 'orange')
ax2.tick_params('y', colors = 'blue')

# Saving the graph
plt.savefig("NewCarsChangeinPriceandMileages.jpg", quality = 50)
plt.show()
```

```python
# Grouping the newer cars per year and plotting the chart against its price

newercarssort.groupby(['ManufactureYear', 'Car Names']).sum().unstack().plot(figsize = (10,10),kind='bar', y='Price', stacked = True, title = "Total Price for New Cars" )

# Saving the graph
plt.savefig("SumPricesofNewCars.jpg", quality = 50)
plt.show()
```

```python
# Comparing the difference between the old cars and newer cars

fig, ax = plt.subplots(2,1, figsize = (30,20))
ax[0].barh(car2018sort["Car Names"], car2018sort['Mileages'])
ax[1].barh(newercarssort["Car Names"], newercarssort['Mileages'])


# Saving the graph
plt.savefig("Differencebetweenoldandnew.jpg", quality = 50)
plt.show()
```
